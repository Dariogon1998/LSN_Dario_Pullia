```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit


N=100
CallPriceDir, CallPriceDirError,PutPriceDir, PutPriceDirError,CallPriceStep, CallPriceStepError,PutPriceStep, PutPriceStepError,  = np.loadtxt("/home/dario/Uni_ubu/lsn/03/data01/data01.dat", usecols=(0,1,2,3,4,5,6,7), delimiter=' ', unpack='true')
x = np.arange(N) 
fig=plt.figure(figsize=(8, 6))

plt.errorbar(x,CallPriceDir, yerr=CallPriceDirError, label="Call Price with Direct method")
plt.errorbar(x,CallPriceStep, yerr=CallPriceStepError,label="Call Price with Step method")

plt.xlabel('#block')
plt.ylabel('Call Price')
plt.axhline(y = 14.97579, color = 'r', linestyle = '-',label="Call Price with Analytical method")
plt.grid(True)
plt.legend()
plt.show()




```

```python
fig=plt.figure(figsize=(8, 6))
plt.errorbar(x,PutPriceDir, yerr=PutPriceDirError, label="Put Price with Direct method")
plt.errorbar(x,PutPriceStep, yerr=PutPriceStepError, label="Put Price with Step method")
plt.xlabel('#block')
plt.ylabel('Put Price')
plt.axhline(y =5.45953, color = 'r', linestyle = '-', label="Put Price with Analytical method")
plt.grid(True)
plt.legend()
plt.show()

```
