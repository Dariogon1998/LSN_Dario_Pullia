### Esercizio 02.1

Cumulo retta $\frac{\pi}{2}(1-x)$ e la funzione stessa. Nel secondo caso ho varianza 0! Giusto!

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def f (x, k):
    return k * np.sqrt(x)


N=100
IntUnif, IntUnifErr, IntLin, IntLinErr,IntCos, IntCosErr  = np.loadtxt("/home/dario/Uni_ubu/lsn/02/data01/data01.dat", usecols=(0,1,2,3,4,5), delimiter=' ', unpack='true')
x = np.arange(N) 
fig=plt.figure(figsize=(12, 8))
plt.errorbar(x,IntUnif, yerr=IntUnifErr)
plt.errorbar(x,IntLin, yerr=IntLinErr)


plt.xlabel('#block')
plt.ylabel('Integral')
plt.axhline(y = 1, color = 'r', linestyle = '-')
plt.grid(True)
plt.show()



```

```python
fig=plt.figure(figsize=(12, 8))
plt.errorbar(x,IntCos, yerr=IntCosErr)
plt.xlabel('#block')
plt.ylabel('Integral')
plt.axhline(y = 1, color = 'r', linestyle = '-')
plt.grid(True)
plt.show()
```

### Esercizio 02.2

```python
Disc, ErrDisc, Cont, ErrCont= np.loadtxt("/home/dario/Uni_ubu/lsn/02/data02/data02.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')
x = np.arange(N+1) 

fig=plt.figure(figsize=(12, 8))
plt.errorbar(x,Cont, yerr=ErrCont,label="Experiment")
plt.xlabel('#block')
plt.ylabel('Integral')

popt, pcov = curve_fit(f, x, Disc)
plt.plot(x,f(x,*popt), color='red', label='Fit')

plt.legend()
plt.show()

```
